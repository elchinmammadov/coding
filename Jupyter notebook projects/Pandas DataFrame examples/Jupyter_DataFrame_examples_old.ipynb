```python
import pandas as pd

salesdf = pd.read_csv('10000_Sales-Records.csv', na_values='n/a', parse_dates=['Order Date', 'Ship Date']) # creates DataFrame from CSV file. 
salesdf.sample(5) # shows 5 random rows in DataFrame. Replaces 'n/a' values with NaN, which tells Python that it's a cell with missing value. Parses values in 'DoB' column as dates

```

```python
salesdf.describe()
```

```python
salesdf[['Order ID', 'Units Sold']] # only shows data for selected columns (i.e. 'Order ID' and 'Units Sold')
```

```python
print(salesdf.at[1, 'Units Sold']) # Access single value by label. In this case, shows value under 'Units Sold' column where row index equals 1 
print(salesdf.iat[1, 8]) # Access single value by index. In this case, shows value where column index = 8 and and row index = 1 
salesdf['Units Sold'].mean() # shows mean of all values under 'Units Sold' column
```

```python
salesdf.iloc[:5] # show first five entries (by index) in DataFrame
```

```python
salesdf = salesdf.sort_values('Units Sold', ascending = False) # sort values in 'Units Sold' column from highest to lowest
salesdf
```

```python
salesdf.sort_index() # sort index of DataFrame (i.e. sort values by index)
salesdf
```

```python
salesdf = salesdf[salesdf['Units Sold'] > 100] # remove all rows where values in Units Sold column is 100 or lower
salesdf = salesdf[salesdf['Order Priority'] == 'L'] # remove all rows where values in Order Priority are not 'L'
salesdf
```

```python
salesdf.loc[salesdf['Country'] == 'Iceland'] # filter rows by condition. In this case, where values under 'Country' column equals 'Iceland'
```

```python
salesdf.loc[salesdf['Country'] == 'Iceland', ['Region', 'Country']].head(5) # filter rows by condition (i.e .where values under 'Country' column equals 'Iceland') and show selected columns (i.e. 'Region' & 'Country') for the first 5 entries in the DataFrame
```

```python
#salesdf[(salesdf['Country'] == 'Iceland') & (salesdf['Item Type'] == 'Snacks') & (salesdf['Units Sold'] > 6000) & salesdf['Sales Channel'].str.startswith('On')] # filter rows using multiple conditions
salesdf[(salesdf['Country'] == 'Iceland') & (salesdf['Item Type'] == 'Snacks') & (salesdf['Units Sold'] > 6000) & salesdf['Sales Channel'].str.startswith('On')][['Country','Item Type','Sales Channel']] # filter rows using multiple conditions and show selected columns
```

```python
salesdf['Units Sold'] = salesdf['Units Sold']/1000 # divide every cell in column 'Units Sold' by 1000
salesdf['Units Sold'] = salesdf['Units Sold']*1000 # multiply every cell in column 'Units Sold' by 1000
salesdf
```

```python
salesdf = salesdf.reset_index() # resets index for DataFrame
salesdf
```

```python
salesdf['Region'].unique() # show non-repeating (i.e. unique) values for Region column
```

```python
salesdf['Country Copy'] = salesdf['Country'].apply(str.upper) # create 'Country Copy' column and populate it with data from 'Country' column using UPPERCASE

salesdf['Total Revenue Scaled'] = salesdf['Total Revenue'].div(1e3) # create 'Total Revenue Scaled' column and populate it with data from 'Total Revenue' column divided by 1 thousand

import numpy as np
salesdf['Empty Column'] = np.NaN # create 'Empty Column' column and populate it with no data (i.e. NaN)
salesdf
```

```python
salesdf = salesdf.drop('Country Copy', 1) # remove 'Country Copy' column
salesdf = salesdf.drop('Total Revenue Scaled', 1) # remove 'Total Revenue Scaled' column
salesdf = salesdf.drop('Empty Column', 1) # remove 'Empty Column' column
salesdf
```

```python
salesdf.rename(columns = {'Order ID': 'Order_ID'}, inplace =True) # rename 'Order ID' column to 'Order_ID' column
salesdf.rename(columns = {'Units Sold': 'Units_Sold'}, inplace =True) # rename 'Units Sold' column to 'Units_Sold' column
salesdf
```

```python
# change values in a pandas DataFrame column based on a condition

salesdf.loc[salesdf.Country == "Cuba", "Country"] = "Kuba" # replaces all 'Cuba' values in 'Country' column with 'Kuba' values
salesdf.loc[salesdf.Order_ID == 534739196, "Country"] = "Kuw8" # replaces value under 'Country' column with 'Kuba' for row where 'Order ID' equals 534739196

import numpy as np
salesdf.loc[salesdf.Order_ID == 947415292, "Country"] = np.NaN # replaces value under 'Country' column with NaN (i.e. blank) for row where 'Order ID' equals 947415292

salesdf
```

```python
salesdf.loc[:,salesdf.isnull().any()] # Select columns containing NaN values
#salesdf.loc[:,salesdf.notnull().any()] # Select columns not containing NaN values
```

```python
salesdf.fillna('Not Applicable') # replaces NaN values with a set variable (i.e. 'Not Applicable')
```

```python
import numpy as np
salesdf.loc[salesdf.Country == "Not Applicable", 'Country'] = np.NaN # replaces 'Non Applicable' values under Country column with 'NaN'
salesdf.dropna() # drop NaN values. In this case, drops row index 316 where value under 'Country' column is set to 'NaN'
```

```python
print(salesdf['Country'].nunique()) # counts non-repeating (i.e. unique) number of countries in Country column
print(salesdf['Country'].count()) # counts only non-null values in Country column
print(salesdf['Country'].size) # counts all values (including null values) in Country column
```

```python
grouped_country_df = salesdf.groupby('Country')
grouped_country_df.describe() # show summary statistics about the data in pandas DataFrame grouped by 'Country'
```

```python
grouped_country_df = salesdf.groupby(by='Country')
grouped_country_df.size() # counts number of rows per country using 'Country' column
```

```python
grouped_region_df = salesdf.groupby('Region').Units_Sold.sum().sort_index() # group/aggregate data in pandas DataTable by a single group variable and a single category. I.e. in this case calculates the total number of units sold by region and sorts the index by region name.
grouped_region_df
```

```python
import plotly_express as px
import pandas as pd

# create geographic maps using plotly express library
# full instructions are here: https://www.youtube.com/watch?v=Oht6cf-Acl0

country_data = px.data.gapminder()
country_data.tail()
```

```python
map_fig = px.scatter_geo(country_data,
                        locations='iso_alpha',
                        projection='orthographic',
                        color='continent',
                        opacity=0.8,
                        hover_name='country',
                        hover_data=['lifeExp', 'pop', 'year']
)
map_fig.show()
```
